```python
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
```

```python
#dataset 2
df2 = pd.read_csv('credit_record.csv')
df2.head()
```

```python
#merged dataset
left_join = pd.merge(df, df2, how="left", on="ID")
left_join
```

```python
left_join.isna().sum()
```

```python
left_join.describe()
```

```python
left_join_duplicated = left_join[left_join['ID'].duplicated(keep=False)]
left_join_duplicated.shape
```

```python
left_join_duplicated
```

```python
left_join_dup = pd.DataFrame(left_join_duplicated['ID'])
left_join_noDup_view = left_join_dup.groupby(list(left_join_dup)).apply(lambda x: tuple(x.index)).tolist()
left_join_noDup_view
```

```python
import random
listIndexRemoved = []
for item in left_join_noDup_view:
    nbDup = len(item)
    indexTupleKept = random.randint(0, nbDup-1)
    for j in range(0, nbDup-1):
        if j != indexTupleKept:
             listIndexRemoved.append(item[j])
```

```python
left_join_cleaned = left_join_duplicated.drop(listIndexRemoved)
left_join_cleaned.duplicated().sum()
```

```python
left_join_cleaned
```

```python
left_join_copy = left_join_cleaned.copy()
left_join_copy.drop(['MONTHS_BALANCE'], axis=1, inplace=True)
```

```python
left_join_copy["STATUS"] = left_join_copy["STATUS"].replace(["1","0","C","X"],"0")
```

```python
left_join_copy["STATUS"] = left_join_copy["STATUS"].replace(["2","3","4","5"],"1")
```

```python
left_join_copy['STATUS'].fillna(7, inplace=True) #dengan catatan sebagai nilai asal untuk mengisi missing values
```

```python
left_join_copy['STATUS'] = left_join_copy['STATUS'].astype('int64')
```

```python
#train dan valid
x = df_train_val.drop(columns=["STATUS"])
y = df_train_val["STATUS"]
```

```python
# train test split
from sklearn.model_selection import train_test_split
train_x, valid_x, train_y, valid_y = train_test_split(x,y, test_size=0.2, random_state=42)
print("Train_x :",train_x.shape)
print("Valid_x :",valid_x.shape) #test validasi
print("Train_y :",train_y.shape)
print("Valid_y :",valid_y.shape)
```

# Handle with Oversampling

```python
from imblearn.over_sampling import SMOTE

rus = SMOTE(random_state=0)
x_resampled,y_resampled = rus.fit_resample(x,y)
```

```python
ax = sns.countplot(y_resampled)
plt.bar_label(ax.containers[0])
plt.title("STATUS", fontsize=18,color="black");
```

```python
train_x, valid_x, train_y, valid_y = train_test_split(x_resampled,y_resampled, test_size=0.2, random_state=42)
print("Train_x :",train_x.shape)
print("Valid_x :",valid_x.shape) #test validasi
print("Train_y :",train_y.shape)
print("Valid_y :",valid_y.shape)
```

```python
#Extra Tree Classifier After Oversampling
from sklearn.ensemble import ExtraTreesClassifier
ETC_model = ExtraTreesClassifier()
clfETC = ETC_model.fit(train_x, train_y)
valid_y_pred = clfETC.predict(valid_x)
from sklearn.metrics import confusion_matrix, classification_report
print('Training Accuracy :', clfETC.score(valid_x, valid_y))  
```

```python
print(classification_report(valid_y,valid_y_pred))
```

```python
from sklearn.metrics import confusion_matrix

sns.heatmap(confusion_matrix(valid_y, clfETC.predict(valid_x)), annot=True, fmt=".0f")
plt.title("Extra Trees Confusion Matrix",fontsize=18, color="black");
```

```python
#Ada Boost Classifier After Oversampling
from sklearn.ensemble import AdaBoostClassifier
ADC_model = AdaBoostClassifier()
clfADC = ADC_model.fit(train_x, train_y)
valid_y_pred = clfADC.predict(valid_x)
from sklearn.metrics import confusion_matrix, classification_report
print('Training Accuracy :', clfADC.score(valid_x, valid_y)) 
```
